### 1. Plot the midnight Temperatures in Eilat as function of time

```python
eilat_data = weather_data[(weather_data["station"] == "אילת 05/2001-07/2025") & (weather_data["date"].dt.hour == 0)]
plt.plot(eilat_data["date"], eilat_data["temp"])
plt.show()
```

### 2. Plot the temperature histogram of the midnight temperatures.

```python
plt.hist(eilat_data["temp"], bins=301)
plt.show()
```
