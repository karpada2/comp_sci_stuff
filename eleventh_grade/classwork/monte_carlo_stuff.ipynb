```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
point_count = 100_000
radius = 2

x_values = np.random.uniform(-radius, radius, point_count)
y_values = np.random.uniform(-radius, radius, point_count)

points_inside = np.sum(x_values**2 + y_values**2 < radius**2)

square_area = (2*radius)**2
circle_area = square_area*(points_inside/point_count)
print("π ≈", circle_area/(radius**2))



ax = plt.gca()
ax.add_patch(plt.Circle((0, 0), radius, fill=False))
plt.plot([-radius, radius, radius, -radius, -radius], [radius, radius, -radius, -radius, radius], color="black")
plt.scatter(x_values, y_values, s=6, alpha=0.02)
plt.xlim(((-radius-0.1), (radius+0.1)))
plt.ylim(((-radius-0.1), (radius+0.1)))
plt.show()
```
