```python
class BinaryNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
    def __str__(self):
        if self != None:
            return f"⬐[{self.value}]⬎"
        else:
            return "[None]"
    
    def get_value(self):
        return self.value
    def set_value(self, value):
        self.value = value

    # value may be node or int, if int changes value of left
    def set_left(self, value):
        if self.left == None:
            if isinstance(value, int):
                self.left = BinaryNode(value)
            elif isinstance(value, BinaryNode):
                self.left = value
            else:
                raise TypeError("value must be BinaryNode or int!")
        else:
            if isinstance(value, int):
                self.left.set_value(value)
            elif isinstance(value, BinaryNode):
                self.left = value
            else:
                raise TypeError("value must be BinaryNode or int!")
    
    def get_left(self):
        return self.left
    
    # value may be node or int, if int changes value of left
    def set_right(self, value):
        if self.right == None:
            if isinstance(value, int):
                self.right = BinaryNode(value)
            elif isinstance(value, BinaryNode):
                self.right = value
            else:
                raise TypeError("value must be BinaryNode or int!")
        else:
            if isinstance(value, int):
                self.right.set_value(value)
            elif isinstance(value, BinaryNode):
                self.right = value
            else:
                raise TypeError("value must be BinaryNode or int!")
    
    def get_right(self):
        return self.right
    

    def get_depth(self):
        left_depth = 0 if self.get_left() == None else self.get_left().get_depth()
        right_depth = 0 if self.get_right() == None else self.get_right().get_depth()

        return max(left_depth, right_depth) + 1

class BinaryTree:
    # value may be either int or BinaryNode
    def __init__(self, value=None):
        self.root = None
        if isinstance(value, int):
            self.root = BinaryNode(value)
        elif isinstance(value, BinaryNode):
            self.root = value

    def get_max_depth(self):
        return self.root.get_depth()
    
    def is_value_in_node(node, value):
        if node == None:
            return False
        elif isinstance(node, BinaryNode):
            if node.get_value() == value:
                return True
            else:
                return BinaryTree.is_value_in_node(node.get_left(), value) or BinaryTree.is_value_in_node(node.get_right(), value)
    
    def is_value_in_tree(self, value):
        BinaryTree.is_value_in_node(self.root, value)
```
