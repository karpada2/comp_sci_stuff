```python
import numpy as np
import matplotlib.pyplot as plt

# data to fit
data = np.genfromtxt('linear_regression_data.csv', delimiter=',', skip_header=1)
x_data=data[:,0]; y_data=data[:,1]

plt.figure(figsize=(8,6))
plt.plot(x_data,y_data,'ro')
plt.xlim(0,140)
plt.ylim(0,140)
plt.title('The input data',fontsize=13)
plt.ylabel('$y$',fontsize=13)
plt.xlabel('$x$',fontsize=13)
plt.grid()
plt.show() 
```

```python
thetas = [0, 1.4]

alpha = 0.000001

def h(x, thetas_in):
    return thetas_in[0] + thetas_in[1] * x

def J(thetas_in):
    return np.mean((h(x_data, thetas_in) - y_data)**2)/2.0

for _ in range(50000):
    temp_thetas = [x for x in thetas]
    temp_thetas[0] = thetas[0] - alpha*(np.mean(h(x_data, thetas) - y_data))
    temp_thetas[1] = thetas[1] - alpha*(np.mean(x_data*(h(x_data, thetas) - y_data)))
    thetas = temp_thetas

print(thetas)
plot_x = np.linspace(0, 140, 1000)
plot_y = h(plot_x, thetas)

plt.plot(plot_x, plot_y)
plt.scatter(x_data, y_data)
plt.show()
```
