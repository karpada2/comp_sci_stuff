```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
```

```python
size = 5
rand_matrix = np.random.rand(5,5)
for i in range(1,size):
    for j in range(0, i+1):
        if (i == j):
            rand_matrix[i][j] = 0
        else:
            rand_matrix[i][j] = -rand_matrix[j][i]

print(rand_matrix)
print(np.array_equal(rand_matrix, rand_matrix.T))
```

```python
def anti_diagonal(size):
    matrix = np.eye(size)
    for i in range(size):
        matrix[i][size - 1 - i] = 1
        matrix[i][i] = 0
    return matrix

print(anti_diagonal(6))
```

```python
n = 10

M = np.zeros((n,n))
M[3:7,3:7] = 1
M[1:3, 4:6] = 7
M[4:6, 7:9] = 5
M

J = anti_diagonal(n)
print(J@M@J)
```
