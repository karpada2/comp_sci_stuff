```python
class Node:
    def __init__(self, value=0, next=None):
        self.value = value
        if next == None or isinstance(next, Node):
            self.next = next
        else:
            raise TypeError("next must be Node!")
    def get_next(self):
        return self.next
    def set_next(self, next):
        if next == None or isinstance(next, Node):
            self.next = next
        else:
            raise TypeError("next must be Node!")
    def get_value(self):
        return self.value
    def set_value(self, value):
        self.value = value
    def __str__(self):
        if self.next == None:
            return f"[{self.value}] --> [None]"
        else:
            return f"[{self.value}] --> " + str(self.get_next())

class LinkedList:
    def __init__(self, root=None):
        self.root = root
    
    def __str__(self):
        return str(self.root)

    def length(self):
        count = 0
        current = self.root
        while current != None:
            count += 1
            current = current.get_next()
        return count
    
    def _get_node(self, index):
        list_length = self.length()
        if index >= list_length or index < 0:
            raise IndexError("index is invalid! (too big or too high)")
        else:
            current = self.root
            for _ in range(index):
                current = current.get_next()
            return current
        
    def get(self, index):
        return self._get_node(index).get_value()
        
    def append(self, value):
        if self.root == None:
            self.root = Node(value)
        else:
            self._get_node(self.length()-1).set_next(Node(value, None))
    
    def remove(self, index):
        if self.root != None:
            if index == 0:
                self.root = self.root.get_next()
            else:
                current_node = self._get_node(index-1)
                current_node.set_next(current_node.get_next().get_next())
        else:
            raise IndexError("List is empty!")
    
    def insert(self, index, value):
        if self.root != None:
            if index == 0:
                self.root = Node(value, self.root)
            else:
                current_node = self._get_node(index-1)
                new_node = Node(value, current_node.get_next())
                current_node.set_next(new_node)
        else:
            if index == 0:
                self.root = Node(value)
            else:
                raise IndexError("index is invalid! List isn't big enough!")
    
    def add_first(self, value):
        self.insert(0, value)
    
    def replace(self, index, value):
        self._get_node(index).set_value(value)
    
    def get_slice(self, start, end, step=1):
        result_list = LinkedList()
        for i in range(start, end, step):
            result_list.append(self.get(i))
        return result_list

# a = Node(17)
# b = Node(7, a)
# c = Node(5, b)

# print(str(c))

lst = LinkedList()
lst.append(1)
lst.append(4)
lst.append(9)
lst.append(16)
lst.add_first(0)

print("original")
print(lst)

lst.remove(0)

print("\n\nafter remove 0")
print(lst)

lst.remove(2)

print("\n\nafter remove 2")
print(lst)

lst.insert(2, 9)

print("\n\nafter insert(2, 9)")
print(lst)
print("\n\nafter slice(1, 3)")
print(lst.get_slice(1, 3))
```
