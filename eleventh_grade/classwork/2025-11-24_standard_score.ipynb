```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
sigma = 7
mew = 163
z = (155-163)/sigma

print(scipy.stats.norm.cdf(z))
```

```python
z1 = (160-mew)/sigma
z2 = (190-mew)/sigma

print(1-(scipy.stats.norm.cdf(z2)-scipy.stats.norm.cdf(z1)))
```

```python
N = 100
n = 1000
series = np.arange(1, 7)

random_means = np.mean(np.random.randint(1, 7, (n, N)), axis=1)

mu = series.mean()
sigma_3 = series.std()/np.sqrt(N)

x = np.linspace(3, 4, 1000)
y = lambda x: (1/(sigma_3*((2*scipy.constants.pi)**0.5)))*np.exp(-((x-mu)**2)/(2*(sigma_3**2)))

plt.hist(random_means, density=True)
plt.plot(x, y(x))
plt.show()
print(np.mean(random_means))
print(np.std(random_means))
```
