```python
import math
import scipy

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __str__(self):
        return f"({self.x}, {self.y})"
    def angle_to(self, otherPoint):
        return math.atan2(otherPoint.y - self.y, otherPoint.x - self.x)


class Student:
    def __init__(self, name, grade):
        self.name = name
        self.grade = grade


class PhysicalMeasurement:
    def __init__(self, value, unit):
        self.value = value
        self.unit = unit
    def __str__(self):
        return f"{self.value} [{self.unit}]"
    

class Experiment:
    def __init__(self, name, repeats):
        self.name = name
        self.repeats = repeats
    def __str__(self):
        return f"Experiment({self.name}, {self.repeats})"


class Vector:
    def __init__(self, lst):
        self.lst = lst
        self.abs_sum = sum([abs(i) for i in self.lst])
    def __str__(self):
        return f"Vector {self.lst}"
    def l1(self):
        return self.abs_sum


class Segment:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
    def __str__(self):
        return f"Segment {str(self.p1)} ----- {str(self.p2)}"
    def length(self):
        return (((self.p2.x - self.p1.x)**2) + ((self.p2.y - self.p1.y)**2))**0.5


class LinearFunction:
    def __init__(self, m, b):
        self.m = m
        self.b = b
    def __str__(self):
        symbol = ("+" if self.b >= 0 else "-")
        return f"f(x) = {self.m}x {symbol} {abs(self.b)}"
    def value(self, x):
        return (self.m*x) + self.b
```

```python
p = PhysicalMeasurement(5, "m/s")
print(p)

segment = Segment(Point(3, 4), Point(5, 10))
print(segment)
print(segment.length())

linearFunction = LinearFunction(2, 5)
print(linearFunction)
print(linearFunction.value(3))
```

```python
class Array:
    def __init__(self, elements):
        self.data = elements
    def get(self, index):
        return self.data[index]

class LimitedArray:
    def __init__(self, capacity, starting_elements=[]):
        self.capacity = capacity
        self.data = starting_elements[:capacity]
        self.size = len(starting_elements)
    def append(self, element):
        able_to_append = self.size < self.capacity
        if able_to_append:
            self.size += 1
            self.data.append(element)
            return True
        else:
            print("capacity reached!")
            return False
    def get(self, index):
        return self.data[index]
    def __str__(self):
        return f"LimitedArray({self.data}, size = {self.size}, capacity = {self.capacity})"


limitedArray = LimitedArray(5)
for i in range(10):
    print(limitedArray.append(i), "\n")
```
