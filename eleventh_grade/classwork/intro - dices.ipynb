### 2. Rolling One Die
You are asked to simulate the experiment of tossing a fair coin.
* Write Python code that simulates 10 coin tosses and prints the sequence of results.
* Extend your program to simulate increase toses until 10,000 and plot a histogram.
* Compare your result with the theoretical probability.
* Rewrite your code so no loops are used (use numpy...)

```python
amount = 1000000
sides = 6
```

```python
import random

results = []

for i in range(amount):
    results.append(random.randint(1, sides))

plt.hist(results, bins=list(range(1, sides+2)), width=0.5)
plt.plot([1, sides+0.5], [amount/sides, amount/sides])
plt.show()
```

### 3. Rolling Two Dice
Consider the experiment of rolling two fair dice.
* Write Python code that simulates a pair of dice rolls and prints thier sums.
* For each roll, calculate the total sum, and plot a histogram 1000 sums.
* Rewrite your code so no loops are used (use numpy...)

```python
def roll_dice(dice_sides, dice_amount, rolls_to_perform):
    results = np.random.randint(1, dice_sides+1, (rolls_to_perform, dice_amount))
    
    return np.sum(results, axis=1)
```

```python
rolls_3 = 1000
amount_3 = 2
dice_sides_3 = 6

dice_rolls = roll_dice(dice_sides_3, amount_3, rolls_3)
plt.hist(dice_rolls, bins=range(amount_3, (amount_3*dice_sides_3)+2), width=0.8)
plt.show()
```
