```python
# 1).
class Particle:
    def __init__(self, position, velocity):
        self.position = position
        self.velocity = velocity
    def get_position(self, time):
        return self.position + (time*self.velocity)

p1 = Particle(0, 2)
p2 = Particle(10, -1)

t = 5

print(p1.get_position(t))
print(p2.get_position(t))
```

```python
# 2).
class LinearFunction:
    def __init__(self, m, b):
        self.m = m
        self.b = b
    
    def describe(self):
        out = "f(x) = "
        if self.m < 0:
            out += "-"
        if self.m != 0:
            if abs(self.m) == 1:
                out += "x"
            else:
                out += f"{self.m}x"
        
        if self.b == 0:
            return out
        
        if self.b < 0:
            out += f" - {-self.b}"
        else:
            out += f" + {self.b}"
        
        return out
    
    def value(self, x):
        return (self.m*x) + self.b


f1 = LinearFunction(2, -3)
f2 = LinearFunction(-1, 5)

x = 4

print(f1.describe())
print(f2.describe())

print(f1.value(x))
print(f2.value(x))
```

```python
# 4).
class Satellite:
    planet_radius = 6371
    count = 0

    def __init__(self, name, height):
        self.name = name
        self.height = height
        Satellite.count += 1
    
    def orbit_radius(self):
        return self.height + Satellite.planet_radius

s1 = Satellite("Hubble", 540)
s2 = Satellite("ISS", 408)

print(s1.orbit_radius())
print(s2.orbit_radius())

print(Satellite.count)
```

```python
# 5).

class CelestialBody:
    def __init__(self, name, distance_from_sun):
        self.name = name
        self.distance_from_sun = distance_from_sun
        self.is_orbiting = True
        # Add the following attributes:
        # 1. name - the body's name
        # 2. distance_from_sun - distance in million km
        # 3. is_orbiting - constant value True (all bodies orbit)
            
    def show_info(self):
        print(f"{self.name} is located {self.distance_from_sun} million km from the Sun.")
        # Print in format:
        # "[name] is located [distance_from_sun] million km from the Sun"
            
    def revolve(self):
        print(f"{self.name} is revolving around the Sun.")
        # Print: "[name] is revolving around the Sun"

class Planet(CelestialBody):
    # Create a class that inherits from CelestialBody
    # Add a method describe() that prints: "[name] is a major planet in the solar system"
    def describe(self):
        print(f"{self.name} is a major planet in the solar system")
    
class Moon(CelestialBody):
    # Create a class that inherits from CelestialBody
    # Add a method describe() that prints: "[name] is a natural satellite orbiting a planet"
    def describe(self):
        print(f"{self.name} is a natural satellite orbiting a planet")
    
class Asteroid(CelestialBody):
    # Create a class that inherits from CelestialBody
    # Add a method describe() that prints: "[name] is a rocky object in the asteroid belt"
    def describe(self):
        print(f"{self.name} is a rocky object in the asteroid belt")
    
# Testing:
mars = Planet("Mars", 228)
europa = Moon("Europa", 778)
ceres = Asteroid("Ceres", 414)

mars.show_info()
mars.revolve()
mars.describe()

print(f"\nEuropa orbiting? {europa.is_orbiting}")
europa.show_info()      
europa.describe()       

print(f"\nCeres details:")
ceres.show_info()       
ceres.describe()        
```

```python
# 6).
class Sequence:
    def __init__(self, name, first_term):
        self.name = name
        self.terms = [first_term]
        self.terms_calculated = 1
    
    def get_term(self, n):
        return
        n -= 1

        if n > self.terms_calculated:
            self.get_term(n)

        if n == self.terms_calculated:
            self.terms.append(
                self.terms[-1] # calculation goes here
                )
            self.terms_calculated += 1
            return self.terms[-1]
        elif n < self.terms_calculated:
            return self.terms[n]
    
    def display_info(self):
        print(f"{self.name}, {self.terms[0]}, {self.terms_calculated}")
    
    def show_first_n_terms(self, n):
        self.get_term(n)
        print(self.terms)
    
    



class ArithmeticSeries(Sequence):
    def __init__(self, name, first_term, d):
        super().__init__(name, first_term)
        self.d = d
    
    def get_term(self, n):
        n -= 1

        if n > self.terms_calculated:
            self.get_term(n)

        if n == self.terms_calculated:
            self.terms.append(
                self.terms[-1] + self.d # calculation goes here
                )
            self.terms_calculated += 1
            return self.terms[-1]
        elif n < self.terms_calculated:
            return self.terms[n]
        
    def get_sum(self, n):
        self.get_term(n)
        return sum(self.terms[:n])

arithmetic = ArithmeticSeries("Ari", 3, 5)
arithmetic.show_first_n_terms(6)
print(arithmetic.get_term(10))
print(arithmetic.get_sum(10))
print("\n")



class GeometricSeries(Sequence):
    def __init__(self, name, first_term, q):
        super().__init__(name, first_term)
        self.q = q
    
    def get_term(self, n):
        n -= 1

        if n > self.terms_calculated:
            self.get_term(n)

        if n == self.terms_calculated:
            self.terms.append(
                self.terms[-1] * self.q # calculation goes here
                )
            self.terms_calculated += 1
            return self.terms[-1]
        elif n < self.terms_calculated:
            return self.terms[n]
    
    def get_product(self, n):
        self.get_term(n)
        product = 1
        for i in range(n):
            product *= self.terms[i]
        return product

geometric = GeometricSeries("Geo", 2, 3)
geometric.show_first_n_terms(5)
print(geometric.get_term(7))
print(geometric.get_product(4))
print("\n")



import math
class FibonacciSeries(Sequence):
    golden_ratio = 1.61803398875
    def get_term(self, n):
        n -= 1

        if n > self.terms_calculated:
            self.get_term(n)

        if n == self.terms_calculated:
            if self.terms_calculated == 1:
                self.terms.append(1)
            else:
                self.terms.append(
                    self.terms[-1] + self.terms[-2] # calculation goes here
                    )
            self.terms_calculated += 1
            return self.terms[-1]
        elif n < self.terms_calculated:
            return self.terms[n]
    
    def is_fibonacci(self, number):
        self.get_term(int(math.log(number, FibonacciSeries.golden_ratio))+2)
        return number in self.terms


fibonacci = FibonacciSeries("Fib", 1)
fibonacci.show_first_n_terms(4)
print(fibonacci.is_fibonacci(21))
print(fibonacci.is_fibonacci(20))
```
