```python
import pandas as pd
import matplotlib.pyplot as plt

weather_data = pd.read_csv('data_202509031156.csv')
```

### Ex. 1 - How wind relates to temperatue?
* Exmaine data first.
* Change columns names to English.
* Use plot.scatter to see how wind and temperatue relates.
* Add axis labels with titles and units.


```python
weather_data.columns = ["station","date", "temperature", "wind"]
weather_data['date'] = pd.to_datetime(weather_data['date'], format="%d-%m-%Y %H:%M")
display(weather_data.head(10))
plt.scatter(weather_data['temperature'], weather_data['wind'], s = 0.5)
plt.xlabel("temperature (C*)")
plt.ylabel("wind speed (m/s)")
plt.show()
```

### Ex. 2 - Compare the relationship between wind and temperature in Eilat versus Tel Aviv
*  Use plot.scatter
*  Add axis labels with titles and units.

```python
eilat_data = weather_data[weather_data["station"] == "אילת 05/2001-07/2025"]
tel_aviv_data = weather_data[weather_data["station"] == "תל-אביב חוף 04/2005-07/2025"]

plt.scatter(eilat_data["temperature"], eilat_data["wind"], alpha=0.5, s=3, label="Eilat")
plt.xlabel("temperature (C*)")
plt.ylabel("wind speed (m/s)")
plt.show()
plt.scatter(tel_aviv_data["temperature"], tel_aviv_data["wind"], alpha=0.5, s=3, label="Tel-Aviv")
plt.xlabel("temperature (C*)")
plt.ylabel("wind speed (m/s)")
plt.show()
```

### Ex. 3 - Comparing the annual maximum temperatures of several stations

* Generate a graph comparing the annual maximum temperatures of Tel Aviv, Jerusalem, Eilat, and Rosh HaNikra from 2005 to 2025.
* Add legend.
* Add axis labels with titles and units.


```python
stations = ['ראש הנקרה 09/2003-07/2025', 'תל-אביב חוף 04/2005-07/2025', 'ירושלים גבעת רם 03/2001-07/2025', 'אילת 05/2001-07/2025']
station_names = ["Rosh Hanikra", "Tel-Aviv", "Jerusalem Giva'at Ram", "Eilat"]
stations_data = [[], [], [], []]

years = range(2005, 2026)

for i in range(len(stations)):
    for year in years:
        stations_data[i].append(weather_data.loc[(weather_data['station'] == stations[i]) & (weather_data['date'].dt.year == year)]['temperature'].max())

for i in range(len(stations)):
    plt.scatter(years, stations_data[i], label=station_names[i])

plt.legend()
plt.show()
```
