```python
from IPython.display import clear_output

dead = "O"
alive = "#"

board_size = (40, 50)

def get_amount_of_alive_around(x, y, board):
    num = 0
    for xPos in range(x-1, x+2):
        for yPos in range(y-1, y+2):
            if xPos >= 0 and yPos >= 0 and xPos < len(board[0]) and yPos < len(board):
                if board[yPos][xPos]:
                    num += 1
    return num

def advance_board(board):
    new_board = [[False]*board_size[0]]*board_size[1]
    for y in range(len(board)):
        for x in range(len(board[0])):
            amount_alive = get_amount_of_alive_around(x, y, board)
            if amount_alive < 2:
                new_board[y][x] = False
            elif amount_alive == 2:
                new_board[y][x] = board[y][x]
            elif amount_alive == 3:
                new_board[y][x] = True
            else:
                new_board[y][x] = False
    return new_board

```
