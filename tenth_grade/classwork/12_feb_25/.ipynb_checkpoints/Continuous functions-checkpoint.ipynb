```python
import matplotlib.pyplot as plt

T_room = 25
k = 2_000_000_000_000_000_000_000_000_000_000
dt = 0.001
data_size = int(10/dt)

T = [0]*data_size
t = [0]*data_size

T[0] = 80

for i in range(data_size-1):
    divT = -k*(T[i] - T_room)
    T[i+1] = T[i] + divT*dt
    t[i+1] = t[i] + dt

plt.plot(t, T)
plt.show()
```

```python
import matplotlib.pyplot as plt

k = 100
r = 4
dt = 0.01
data_size = int(10/dt)

seeds = [10, 80, 100, 120, 140]

P = [0]*data_size
t = [0]*data_size


for seed in seeds:
    P[0] = seed

    for i in range(data_size-1):
        divP = r * P[i] * (1 - (P[i]/k))
        P[i+1] = P[i] + divP*dt
        t[i+1] = t[i] + dt

    plt.plot(t, P, marker="o", label=f"{seed=}")

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

a = 0.2
b = 3
c = 8
dt = 0.001
data_size = int(2/dt)

seeds = [0.05, 2, 4, 6, 7.95, 8, 8.0000001]

x = [0]*data_size
t = [0]*data_size


for seed in seeds:
    x[0] = seed

    for i in range(data_size-1):
        divX = a*x[i]*(b-x[i])*(c-x[i])
        x[i+1] = x[i] + divX*dt
        t[i+1] = t[i] + dt

    plt.plot(t, x, marker="o", label=f"{seed=}")

plt.show()

# fixed points: 3, 8
```
