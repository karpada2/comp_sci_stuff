```python
import matplotlib.pyplot as plt
import time

start_time = time.time()
data_size = 200_000


x0 = 0.6
y0 = 0.6
a = 1.4
b = 0.3

x_func = lambda x, y: y - a*(x**2) + 1
y_func = lambda x, y: b*x

x = [0]*data_size
y = [0]*data_size
t = [0]*data_size

x[0] = x0
y[0] = y0

for i in range(data_size-1):
    x[i+1] = x_func(x[i], y[i])
    y[i+1] = y_func(x[i], y[i])
    t[i+1] = t[i] + 1

plt.plot(t, x, marker="o")
plt.plot(t, y, marker="o")
plt.grid()
plt.show()

plt.scatter(x, y)
plt.grid()
plt.show()
print(f"Time to finish = {time.time() - start_time} sec.")
```
