```python
import turtle
import functools


turtle.Screen().getcanvas().winfo_toplevel().attributes("-fullscreen", True)


def go_to_without_drawing():
    turtle.penup()
    pos = stack.pop()
    turtle.setposition(pos[0])
    turtle.setheading(pos[1])
    turtle.pendown()

def insert_turtle_pos():
    stack.append((turtle.pos(), turtle.heading()))

stack = []
def draw_this(text):
    moves = {
        "A" : functools.partial(turtle.forward, 50),
        "B" : functools.partial(turtle.forward, 50),
        "+" : functools.partial(turtle.left, 22.5),
        "-" : functools.partial(turtle.right, 22.5),
        "[" : insert_turtle_pos,
        "]" : go_to_without_drawing
    }
    for i in text:
        if i in moves:
            moves[i]()

def advance(text):
    laws = {
        "A" : "B-[[A]+A]+B+[+BA]-A",
        "B" : "BB",
        "+" : "+",
        "-" : "-",
        "[" : "[",
        "]" : "]"
    }
    new_text = ""
    for i in text:
        if i in laws:
            new_text += laws[i]
    return new_text

def do_recursive(text, depth):
    if depth == 0:
        return text
    else:
        new_text = advance(text)
        return do_recursive(new_text, depth-1)

turtle.Screen().yscale = 0.25
turtle.Screen().xscale = 0.25
turtle.tracer(0)
print(do_recursive("A", 5))
draw_this(do_recursive("A", 5))
print(stack)
turtle.exitonclick()
turtle.done()
```
