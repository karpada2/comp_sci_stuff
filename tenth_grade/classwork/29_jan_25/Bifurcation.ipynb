```python
import matplotlib.pyplot as plt

data_size = 500
res = 8000
N = 200

final = [[0]*N]*res
x = [0]*data_size
t = [0]*data_size

x[0] = 0.6
a = 2
b = 4

r = [0]*res

for j in range(res):
    r[j] = a + (b-a)/res*j
    
    for i in range(data_size - 1):
        x[i+1] = r[j]*(x[i])*(1-x[i])
        t[i+1] = t[i] + 1
    
    final[j] = x[-N:]

plt.scatter([[r[i]]*N for i in range(res)], final, s=0.3)
plt.show()
```

```python
def repeats_in_list(lst: list, tolerance: float = 5) -> int:
    seen = []
    num = 0
    for i in lst:
        if not round(i, tolerance) in seen:
            num += 1
            seen.append(round(i, tolerance))
    return num

def get_r_values_with_period(values, r_values, period):
    accepted_r_values = []
    
    for i in range(len(values)):
        if repeats_in_list(values[i], 15) == period:
            accepted_r_values.append(r_values[i])

    return accepted_r_values
```

```python
import matplotlib.pyplot as plt
import math

data_size = 500
res = 10000
N = 50

final = [[0]*N]*res
x = [0]*data_size
t = [0]*data_size

x[0] = 0.6
a = 0
b = 4

r = [0]*res

for j in range(res):
    r[j] = a + (b-a)/res*j
    
    for i in range(data_size - 1):
        x[i+1] = r[j]*math.sin((math.pi*x[i])/2.0)
        t[i+1] = t[i] + 1
    
    final[j] = x[-N:]

plt.scatter([[r[i]]*N for i in range(res)], final, s=0.3)
plt.show()
```
