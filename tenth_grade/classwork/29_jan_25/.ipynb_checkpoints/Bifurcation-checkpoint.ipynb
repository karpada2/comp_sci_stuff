```python
import matplotlib.pyplot as plt

data_size = 500
res = 2000
N = 50

final = [[0]*N]*res
x = [0]*data_size
t = [0]*data_size

x[0] = 0.5
a = 3.95
b = 4

r = [0]*res

for j in range(res):
    r[j] = a + (b-a)/res*j
    
    for i in range(data_size - 1):
        x[i+1] = r[j]*x[i]*(1-x[i])
        t[i+1] = t[i] + 1
    
    final[j] = x[-N:]

plt.scatter([[r[i]]*N for i in range(res)], final, s=0.3)
```
