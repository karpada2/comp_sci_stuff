```python
import matplotlib.pyplot as plt
import math
plt.rcParams["figure.figsize"] = (20, 10)
# 4. Algorithm for Histogram of Grades

def tuple_to_range_string(tup):
    return f"{tup[0]}-{tup[1]}"

data = [9,10,10,20,30,40,40,40,50,60,60,60,60,60,60,100,100]

min_value = 0
max_value = 100
bin_amount = 20

bins = [0]*bin_amount
ranges = [()]*bin_amount

step = (max_value-min_value)/bin_amount

for i in range(bin_amount-1):
    ranges[i] = (math.ceil(min_value + step*i), math.ceil(min_value + step*(i+1) - 1))

ranges[-1] = (math.ceil(min_value + step*(bin_amount-1)), max_value)

for num in data:
    for i in range(len(ranges)):
        if num >= ranges[i][0] and num <= ranges[i][1]:
            bins[i] += 1


plt.bar([i for i in range(bin_amount)], bins)
plt.xticks([i for i in range(bin_amount)], [tuple_to_range_string(i) for i in ranges])
plt.show()
```
