```python
import matplotlib.pyplot as plt

a = 1
b = 0.5
c = 0.2
d = 0.6
R_0 = 8
F_0 = 4
dt = 0.0001
total_time = 2000
data_size = int(total_time/dt)

R = [0]*data_size
F = [0]*data_size
t = [0]*data_size

R[0] = R_0
F[0] = F_0

for i in range(data_size-1):
    dR = a*R[i] - b*R[i]*F[i]
    dF = c*R[i]*F[i] - d*F[i]
    R[i+1] = R[i] + dR*dt
    F[i+1] = F[i] + dF*dt
    t[i+1] = t[i] + dt

plt.plot(t, R)
plt.show()
plt.plot(t, F)
plt.show()
plt.plot(R, F)
plt.show()
```

```python
import matplotlib.pyplot as plt

sigma = 10
beta = 2.66
rho = 8

x_0 = 1
y_0 = 0
z_0 = 0

dt = 0.0001
total_time = 2
data_size = int(total_time/dt)

x = [0]*data_size
y = [0]*data_size
z = [0]*data_size
t = [0]*data_size

x[0] = x_0
y[0] = y_0
z[0] = z_0

for i in range(data_size-1):
    dx = sigma*(y[i]-x[i])
    dy = x[i]*(rho-z[i]) - y[i]
    dz = x[i]*y[i] - beta*z[i]
    x[i+1] = x[i] + dx/dt
    y[i+1] = y[i] + dy/dt
    z[i+1] = z[i] + dz/dt
    t[i+1] = t[i] + dt

plt.plot(t, z)
plt.show()
plt.plot(t, y)
plt.show()
plt.plot(t, z)
plt.show()
```
