```python
import matplotlib.pyplot as plt

a = 1
b = 0.5
c = 0.2
d = 0.6
R_0 = 8
F_0 = 4
dt = 0.001
data_size = int(10/dt)

R = [0]*data_size
F = [0]*data_size
t = [0]*data_size

R[0] = R_0
F[0] = F_0

for i in range(data_size-1):
    divR = a*R[i] - b*R[i]*F[i]
    divF = c*R[i]*F[i] - d*F[i]
    R[i+1] = R[i] + divR*dt
    F[i+1] = F[i] + divF*dt
    t[i+1] = t[i] + dt

plt.plot(t, R)
plt.show()
plt.plot(t, F)
plt.show()
plt.plot(R, F)
plt.show()
```
