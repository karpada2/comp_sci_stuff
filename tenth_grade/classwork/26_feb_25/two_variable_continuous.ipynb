```python
import matplotlib.pyplot as plt

a = 1
b = 0.5
c = 0.2
d = 0.6
R_0 = 8
F_0 = 4
dt = 0.0001
total_time = 2000
data_size = int(total_time/dt)

R = [0]*data_size
F = [0]*data_size
t = [0]*data_size

R[0] = R_0
F[0] = F_0

for i in range(data_size-1):
    dR = a*R[i] - b*R[i]*F[i]
    dF = c*R[i]*F[i] - d*F[i]
    R[i+1] = R[i] + dR*dt
    F[i+1] = F[i] + dF*dt
    t[i+1] = t[i] + dt

plt.plot(t, R)
plt.show()
plt.plot(t, F)
plt.show()
plt.plot(R, F)
plt.show()
```

```python
import matplotlib.pyplot as plt

sigma = 10
beta = 2.66
rho = 8

x_0 = 1
y_0 = 0
z_0 = 0

dt = 0.0001
total_time = 2
data_size = int(total_time/dt)

x = [0]*data_size
y = [0]*data_size
z = [0]*data_size
t = [0]*data_size

x[0] = x_0
y[0] = y_0
z[0] = z_0

for i in range(data_size-1):
    dx = sigma*(y[i]-x[i])
    dy = x[i]*(rho-z[i]) - y[i]
    dz = x[i]*y[i] - beta*z[i]
    x[i+1] = x[i] + dx/dt
    y[i+1] = y[i] + dy/dt
    z[i+1] = z[i] + dz/dt
    t[i+1] = t[i] + dt

plt.plot(t, z)
plt.show()
plt.plot(t, y)
plt.show()
plt.plot(t, z)
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

def plot_with_arrows(x, y, width_in = 0.015, **kwargs):
    for i in range(1, len(x)):
        plt.arrow(x[i-1], y[i-1], x[i]-x[i-1], y[i]-y[i-1], head_width=width_in, zorder=1000, length_includes_head = True)

def get_cobweb(function, x0, repeats):
    out_x = [x0]
    out_y = [function(x0)]
    
    for i in range(1, int(repeats)+1, 2):
        out_x.append(out_y[i-1])
        out_y.append(out_y[i-1])
        out_x.append(out_x[i])
        out_y.append(function(out_x[i]))
    return [x0] + out_x,[0] + out_y



f = lambda x: (1+x)**0.5

x_values = [i*0.01 for i in range(-1000, 1001)]
y_values = list(map(f, x_values))

plt.plot(x_values, y_values, label="func")
plt.plot(x_values, x_values, label="linear")

x, y = get_cobweb(f, 0, 10)
plot_with_arrows(x, y)

plt.legend()
plt.xlim(-1, 10)
plt.ylim(0, 4)
plt.grid()
```
