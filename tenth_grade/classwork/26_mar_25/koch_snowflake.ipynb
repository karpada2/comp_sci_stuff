```python
import math
import matplotlib.pyplot as plt

def koch_base_case(x_values, y_values):
    A = (x_values[0], y_values[0])
    B = (x_values[1], y_values[1])
    dx = B[0] - A[0]
    dy = B[1] - A[1]

    C = (A[0] + (1/3) * dx, A[1] + (1/3) * dy)
    D = (C[0] + (1/3) * dx, C[1] + (1/3) * dy)

    E = (A[0] + (1/2) * dx, A[1] + (1/2) * dy)

    P = (E[0] - math.sqrt(1/12) * dy, E[1] + math.sqrt(1/12) * dx)
    
    return_x = [A[0], C[0], P[0], D[0], B[0]]
    return_y = [A[1], C[1], P[1], D[1], B[1]]
    return return_x, return_y



def koch_snowflake(x_values, y_values, depth):
    if depth == 0:
        return x_values, y_values
    
    next_x, next_y = koch_snowflake(x_values, y_values, depth-1)
    new_x = []
    new_y = []
    for i in range(len(next_x)-1):
        temp_x, temp_y = koch_base_case(next_x[i:i+2], next_y[i:i+2])
        new_x.extend(temp_x)
        new_y.extend(temp_y)
    return new_x, new_y
        




x = [0, 0, 1, 1, 0]
y = [0, 1, 1, 0, 0]
plt.plot(x, y)
plt.fill(x, y)
plt.show()

x, y = koch_snowflake(x, y, 4)

plt.plot(x, y)
plt.fill(x, y)

plt.show()
```

```python
def l_system(text, depth):
    if depth == 0:
        return text
    
    new_text = ""
    for i in text:
        new_text += ("a" if i=="b" else "ab")
    return l_system(new_text, depth-1)

seed = "a"
for i in range(5):
    print(l_system(seed, i))
```
