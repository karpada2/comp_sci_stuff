### 1. Simple Line Plot

```python
import matplotlib.pyplot as plt

# Create data using list comprehension
f = lambda x: 2*x**2 + 2*x
x = [0.2 * i for i in range(51)]
y = list(map(f, x))

# Create the plot
plt.plot(x, y)
plt.title('My First Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 2. Create several graphs on the same plot

```python
import matplotlib.pyplot as plt

# Create data using list comprehension
f = lambda x: 2*x**2 + 2*x
x = [0.2 * i for i in range(51)]
y = list(map(f, x))

# Create the plot

plt.plot(
    x,
    y,
    color="coral",
    linestyle="-.",
    linewidth=2.5,
    marker="",
    markersize=10,
)

# gridlines
plt.grid()

# define plot size
plt.rcParams["figure.figsize"] = (15, 10)  # in inches


# axis titles
plt.xlabel("x", fontname="Arial", fontsize=25, fontweight="bold", color="navy")  # x-axis label
plt.ylabel("y", fontname="Arial", fontsize=25, fontweight="bold", color="navy")  # y-axis label

# axes ranges
plt.xlim(0, 10)

# plot title - the parameter y positions the title slightly above the plot frame
plt.title(
    "My first plot",
    fontname="Arial",
    fontsize=28,
    fontweight="bold",
    color="navy",
    y=1.05,
)


plt.show()
# save as image
# plt.savefig('pics/myPlot.jpg')
```

### 3. Advanced Plot Editting

```python
from matplotlib import pyplot as plt

f1 = lambda x: 0.2 * x ** 2
f2 = lambda x: 2 * x
f3 = lambda x: 15.0 ** (1 - 0.1 * x)

x = [0.2 * i for i in range(51)]
y1 = list(map(f1, x))
y2 = list(map(f2, x))
y3 = list(map(f3, x))

# Create the plot
plt.plot(x, y1, label="Parabola")
plt.plot(x, y2, label="Linear")
plt.plot(x, y3, label="Exponential")

 
plt.title('My First Plot')
plt.xlabel('x')
plt.ylabel('y')

# legend
plt.legend(loc="upper left")


plt.show()
```

## 4. Create subplots

```python
from matplotlib import pyplot as plt

# prepare x-data
x = [0.2 * i for i in range(51)]

# prepare y-data
f1 = lambda x: 0.2 * x * x
f2 = lambda x: 2 * x
f3 = lambda x: 15 ** (1 - 0.1 * x)
f4 = lambda x: 7.5 + x - 0.1 * x * x
y1 = list(map(f1, x))
y2 = list(map(f2, x))
y3 = list(map(f3, x))
y4 = list(map(f4, x))

# define plot size
plt.rcParams["figure.figsize"] = (15, 10)  # in inches

# change font size for tick labels
plt.rcParams["xtick.labelsize"] = "x-large"
plt.rcParams["ytick.labelsize"] = "x-large"

# create subplots
fig,ax = plt.subplots(nrows=2,ncols=2)
((ax1, ax2), (ax3, ax4)) = ax  # unpack axes

# plot graphs

# graph 1
ax1.plot(
    x,
    y1,
    color="coral",
    linestyle="-",
    linewidth=1.5,
    marker="o",
    markersize=7,
    label="parabola",
)
ax1.set_ylabel("y", fontname="Arial", fontsize=18, fontweight="bold", color="navy")
ax1.set_title("first plot", fontname="Arial", fontsize=24, fontweight="bold", color="black")
ax1.set_xlim(0, 10)
ax1.set_ylim(0,)
ax1.grid()
ax1.legend()

# graph 2

ax2.plot(
    x,
    y2,
    color="teal",
    linestyle="--",
    linewidth=3.5,
    marker="",
    markersize=10,
    label="straight line",
)
ax2.set_title("second plot", fontname="Arial", fontsize=24, fontweight="bold", color="black")
ax2.set_xlim(0, 10)
ax2.set_ylim(0,)
ax2.grid()
ax2.legend()

# graph 3

ax3.plot(
    x,
    y3,
    color="crimson",
    linestyle="-.",
    linewidth=3.5,
    marker="",
    markersize=10,
    label="exponential",
)

ax3.set_xlabel("x", fontname="Arial", fontsize=18, fontweight="bold", color="navy")
ax3.set_ylabel("y", fontname="Arial", fontsize=18, fontweight="bold", color="navy")
ax3.set_title("third plot", fontname="Arial", fontsize=24, fontweight="bold", color="black")
ax3.set_xlim(0, 10)
ax3.set_ylim(0,)
ax3.grid()
ax3.legend()

# graph 4

ax4.plot(
    x,
    y4,
    color="yellowgreen",
    linestyle=":",
    linewidth=4.5,
    marker="",
    markersize=10,
    label="inverse parabola",
)
ax4.set_xlabel("x", fontname="Arial", fontsize=18, fontweight="bold", color="navy")
ax4.set_title("fourth plot", fontname="Arial", fontsize=24, fontweight="bold", color="black")
ax4.set_xlim(0, 10)
ax4.set_ylim(0,)
ax4.grid()
ax4.legend()

# always include
plt.tight_layout()
plt.show()
```
